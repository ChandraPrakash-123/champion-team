```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ipl_data=pd.read_csv('Downloads/ipldata.csv')
ipl_data.head()
```

```python
ipl_data.shape
```

```python
ipl_data.isnull().sum()
```

```python
ipl_df=ipl_data.drop('umpire3',axis=1,inplace=True)
```

```python
ipl_df=ipl_data.drop('season',axis=1,inplace=True)
```

```python
ipl_data.columns
```

```python
ipl_data.dtypes
```

```python
ipl_data['date']=pd.to_datetime(ipl_data['date'])
ipl_data['year']=ipl_data['date'].dt.year
ipl_data['month']=ipl_data['date'].dt.month
ipl_data['day']=ipl_data['date'].dt.day
```

```python
ipl_data.head()
```

```python
ipl_data['city']=ipl_data['city'].fillna('unknown')
```

```python

sorted_data = ipl_data.sort_values(by='year', ascending=False)
counts = sorted_data['winner'].value_counts()
print(counts)
```

```python
plt.figure(figsize=(10,8))
plt.bar(counts.index, counts)
plt.xlabel('Winner')
plt.ylabel('Count')
plt.title('Counts of IPL Winners')
plt.xticks(rotation=90)
plt.show()
```

```python
winner_df=ipl_data.groupby('year')[['winner','year']]
winner_df.head()
```

```python

winner_df = ipl_data.groupby('year')[['winner', 'year', 'toss_winner']].head()
winner_df['winner'] = winner_df['toss_winner']
tw = winner_df['winner']
counts = tw.value_counts()
print(counts)
```

```python
plt.figure(figsize=(10,8))
plt.bar(counts.index, counts)
plt.xlabel('toss Winner')
plt.ylabel('winning times')
plt.title('toss winner vs Winners')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8,8))
counts = ipl_data['toss_winner'].value_counts()
labels = counts.index
sizes = counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
